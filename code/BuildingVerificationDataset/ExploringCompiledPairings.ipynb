```python
import os
import sys
import platform
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
IMAGE_DIR = "/notebooks/data/LFWA+/COMPILED/compiled_lfwa+"
BASE_DIR = "/notebooks/data/LFWA+/COMPILED"
ALL_PAIRS = "compiled_lfwa+_all_pairs.csv"
LIGHT_PAIRS = "compiled_lfwa+_light_pairs.csv"
DARK_PAIRS = "compiled_lfwa+_dark_pairs.csv"
```

```python
all_pairs =  pd.read_csv("{}/{}".format(BASE_DIR, ALL_PAIRS))
light_pairs =  pd.read_csv("{}/{}".format(BASE_DIR, LIGHT_PAIRS))
dark_pairs =  pd.read_csv("{}/{}".format(BASE_DIR, DARK_PAIRS))
```

```python
light_pairs.info()
```

```python
random_light_matching_sample = light_pairs[light_pairs["y_match_per_group"] == 1].sample(1).gid.item()
```

```python
fig, axes = plt.subplots(
    nrows=1, 
    ncols=2, 
    figsize=(5, 5),
    subplot_kw={'xticks': [], 'yticks': []}
)

for idx, (index, row) in enumerate(light_pairs[light_pairs["gid"] == random_light_matching_sample].iterrows()):
    img = plt.imread("{}/{}".format(IMAGE_DIR, row.file_path))
    axes[idx].imshow(img)
    axes[idx].set_title(row["name"])

plt.tight_layout()
plt.show()
plt.savefig('MatchingLight.png', bbox_inches='tight')
```

```python
random_light_non_matching_sample = light_pairs[light_pairs["y_match_per_group"] == 0].sample(1).gid.item()
```
